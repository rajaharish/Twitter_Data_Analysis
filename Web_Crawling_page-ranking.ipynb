```python
# %load udemy/udemy/items.py

# Define here the models for your scraped items
#
# See documentation in:
# http://doc.scrapy.org/en/latest/topics/items.html

import scrapy
from scrapy.item import Item, Field


class UdemyItem(scrapy.Item):
    # define the fields for your item here like:
    # name = scrapy.Field()
    title = Field()
    link = Field()
    parent= Field()

```

```python
# %load udemy/udemy/spiders/udemy_spider.py
from scrapy.spider import BaseSpider
from scrapy.selector import HtmlXPathSelector
from udemy.items import UdemyItem

class MySpider(BaseSpider):
    name = "udemy"
    allowed_domains = ["www.udemy.com"]
    start_urls = ["https://www.udemy.com/"]

    def parse(self, response):
        for sel in response.xpath('//ul/li'):
            item = UdemyItem()
            item['title'] = sel.xpath('a/span/text()').extract()
            item['link'] = sel.xpath('a/@href').extract()
            item['parent'] = response.url
            yield item
```

```python
import xlrd

wb = xlrd.open_workbook('udemy/items.xls')

sh = wb.sheet_by_index(0)
link = sh.col_values(0) 
parent = sh.col_values(1)
```

```python
completelist=zip(link, parent)
```

```python
import re
import networkx as nx
g = nx.DiGraph()
g.add_edges_from(completelist) 
```

```python
print g.number_of_nodes()
```

```python
print g.number_of_edges()
```

```python
pr = nx.pagerank(g, alpha=0.9)
```
