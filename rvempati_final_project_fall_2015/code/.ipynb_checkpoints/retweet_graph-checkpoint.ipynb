# Calculating the Favorite Url's and graph construction for retweet relationship

```python
import json 
import re
import pandas as pd
import twitter_text
import nltk

tweets_data_path=[]
tweets_data_path.append('data/trumptwitterfeed_dec7_1.txt')
tweets_data_path.append('data/trumptwitterfeed_dec7_2.txt')
tweets_data_path.append('data/trumptwitterfeed_dec8_1.txt')
tweets_data_path.append('data/trumptwitterfeed_dec9_1.txt')
tweets_data_path.append('data/trumptwitterfeed_dec9_2.txt')
tweets_data_path.append('data/trumptwitterfeed_dec10_1.txt')
tweets_data_path.append('data/trumptwitterfeed_dec11_1.txt')


tweets_data = []
for each in tweets_data_path:
    tweets_file = open(each, "r")
    for line in tweets_file:
        try:
            tweet = json.loads(line)
            tweets_data.append(tweet)
        except:
            continue

entities= []

url_retweet_count={}
for tweet in tweets_data:   
    try:
        extractor = twitter_text.Extractor(tweet['text'])
        for url in extractor.extract_urls_with_indices():
            u=str(url['url'])
            entities.append(url)
            if 'retweeted_status' in tweet:
                if u in url_retweet_count:
                    url_retweet_count[u]=int(url_retweet_count[u])+tweet['retweeted_status']['retweet_count']    
                else:
                    url_retweet_count[u]=int(tweet['retweeted_status']['retweet_count'])               
    except:
        continue
print "entities len ,",len(entities)
print "dic len ,",len(url_retweet_count)
```

```python
import operator
sorted_x = sorted(url_retweet_count.items(), key=operator.itemgetter(1))
```

```python
freq=[]
freq=sorted_x[-10:]
for each in freq:
    print "url ",each[0], "Retweet count : ", each[1]
```

```python
import json 
import re
import pandas as pd
import twitter_text
import nltk

#tweets_data_path = 'data/hillarytwitterfeed_dec8_1.txt'

#tweets_data_path = 'data/electiontwitterfeed_dec12_1.txt'

tweets_data_path = 'data/hillarytwitterfeed_dec6_2.txt'



tweets_data = []
tweets_file = open(tweets_data_path, "r")
for line in tweets_file:
    try:
        tweet = json.loads(line)
        tweets_data.append(tweet)
    except:
        continue
```

```python
import networkx as nx
import re
g = nx.DiGraph()


def get_rt_sources(tweet):
    rt_patterns = re.compile(r"(RT|via)((?:\b\W*@\w+)+)", re.IGNORECASE)
    return [ source.strip()
        for tuple in rt_patterns.findall(tweet)
            for source in tuple
                if source not in ("RT", "via") ]

for tweet in tweets_data: 
    try:
        rt_sources = get_rt_sources(tweet["text"])
        if not rt_sources: continue
        for rt_source in rt_sources:
            g.add_edge(rt_source, tweet['user']['screen_name'], {"tweet_id" : tweet["id"]})
            #g.add_edge(tweet['user']['screen_name'], rt_source, {"tweet_id" : tweet["id"]})
    except:
        continue
```
